### part b)

```python
cities_set = set(array1[1:, 1])
```

```python
cities_sumation = {}
for city in cities_set:
  print(np.where(array1[1:, 1] == city)[0])
  for i in np.where(array1[1:, 1] == city)[0]:
    if array1[i + 1][2] == 'temperature':
      continue
    if city in cities_sumation.keys():
      cities_sumation.update({city : cities_sumation.get(city) + float(array1[i + 1][2])})
    else:
      cities_sumation.update({city : float(array1[i + 1][2])})
    
cities_sumation
```

```python
def cities_count(city):
  return len(np.where(array1[1:, 1] == city)[0])
```

```python
avereges = {}
for city in cities_sumation.keys():
  total = cities_sumation.get(city)
  city_count = cities_count(city)
  avereges.update({city : total / city_count})
avereges
```

```python
for city in cities_set:
  max = array1[np.where(array1[1: , 1] == city)[0][0] + 1][2]
  for i in np.where(array1[1: , 1] == city)[0]:
    print(array1[i + 1][2])
    if array1[i + 1][2] > max:
      max = i
  print(f"city name: {city} max temprature : {max}")
```

```python
import pandas as pd
```

```python
df = pd.read_csv('tempratures.csv')
df.shape
```

### part b)

```python
# using pandas.groupby()
groups = df.groupby('city')
for group in groups.groups:
  print(f"city: {group} median: {groups.get_group(group).median()}")
```

### part c)

```python
# calculate minimum and maximum temp for each city
for group in groups.groups:
  print(f"city: {group} range: {groups.get_group(group)['temperature'].max() - groups.get_group(group)['temperature'].min()}")
```

```python
df1 = pd.read_csv('tempratures.csv')
df2 = pd.read_csv('cities.csv')
```

```python
res = df1.merge(df2, how = 'inner', on = 'city')
res
```

```python
is_above = len(res) * [0]
res['is_above'] = is_above
res
```

```python
res.iloc[1]['is_above'] = 1
```

```python
res
```

```python
groups = res.groupby('city')
for i in range(len(res)):
  if res.iloc[i]['temperature'] > groups.get_group(res.iloc[i].city).temperature.mean():
    is_above[i] = 1
res['is_above'] = is_above
```
